```python
import vectorbt as vbt
```

```python
dados = vbt.BinanceData.download('BTCBRL', start='2022-01-24', interval='5m').get()
```

```python
dados
```

```python
prices = dados['Close'] 
media_rapida = vbt.MA.run(prices, 9)
media_lenta = vbt.MA.run(prices, 21)
```

```python
entradas = media_rapida.ma_crossed_above(media_lenta)
saidas = media_rapida.ma_crossed_below(media_lenta)
```

```python
pf = vbt.Portfolio.from_signals(prices, entradas, saidas)
```

```python
pf.plot().show()
```

```python
pf.stats()
```
